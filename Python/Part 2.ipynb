## CNN - Project - Part 2

- Importing images

```python
from skimage.io import imread_collection

#Path 
col_dir = 'Images/*.jpg'

#Creating a collection with the available images
col = imread_collection(col_dir)
```

- Displaying images

```python
# Visualizing the first 15 images in the dataset 

# Visualization
import matplotlib.pyplot as plt 
import seaborn as sns

# Number of Images we will see
N=15  

plt.figure(figsize=(N, 1))
for i in range(N):
    plt.subplot(1, N, i+1)
    plt.imshow(col[i], cmap="gray")  # greens, reds, blues, rgb
    plt.axis('off')
plt.show()
```

- Displaying annotations

- Displaying images vs annotations 

```python
# Visualizing the first 15 images in the dataset and their labels

# Visualization
import matplotlib.pyplot as plt 
import seaborn as sns

# Number of images we will see
N=15  

plt.figure(figsize=(N, 1))
for i in range(N):
    plt.subplot(1, N, i+1)
    plt.imshow(col[i], cmap="gray")  # greens, reds, blues, rgb
    plt.title("{}".format(Label_Make[i]))
    plt.axis('off')
plt.show()
```
